```python
# Independent and Depedent Variables
```

```python
x=data.drop("target",axis=1)
y=data["target"]
```

```python
# Splitting of Dataset into train and test
```

```python
#splitting the data into trainning and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
log.fit(x_train,y_train)

```

```python
y_pred1 = log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred1)


```

```python
labels = np.unique(y_test)     # Get unique class label
cm_df = pd.DataFrame(cm, index=labels, columns=labels)
```

```python
labels=np.unique(y_test)      # Get  unique class label
cm_df = pd.DataFrame(cm, index=labels, columns=labels)
```

```python
# Plot confusion matrix using seaborn
plt.figure(figsize=(6,4))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Purples', linewidths=1, linecolor='black')


plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confussion Matrix")
plt.show()
```
