```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = x*2
```

```python
y
```

```python
#Line Chart
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#Bar Chart
```

```python
plt.bar(x,y,color=["#5B1363","#E49CEC"],edgecolor="black")
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#Scatter Plot
```

```python
a=(1,6,4,0,2,8,9,7,4,0)
b=(21,43,57,68,90,18,79,73,88,39)
plt.scatter(a,b,color="#B86319",s=150)
plt.title("Scatter Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#Pie Chart
```

```python
students = [45, 30, 15, 10]
batches = ['Batch 1', 'Batch 2', 'Batch 3', 'Batch 4']

plt.figure(figsize=(6,6))
plt.pie(
    students, 
    labels=batches, 
    autopct="%1.1f%%", 
    startangle=90, 
    colors=["#8B1EB3", "#F2460D", "#F4930B", "#6DCBEE"]
)

plt.title("Student Distribution in Batches 1–4")
plt.axis('equal')
plt.show()

        
```

```python
# Histogram
```

```python
a = [3,5,6,7,1,4,9,3,0,2,
     2,7,4,4,7,2,3,7,1,8,
     1,9,3,1,2,8,6,5,1,8]

# Create Histogram
plt.hist(a, bins=8, color='#0F957E' , edgecolor='black')

#Labels and title
plt.title("Histogram of Data")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Add Grid
plt.grid(axis='y' , linestyle='--', alpha=0.7)

# Show Plot
plt.show()
```
